```python
sns.kdeplot(df.loc[df['Target']==1].ID_IDADE, color='b', shade=True, Label= "CHURN")
sns.kdeplot(df.loc[df['Target']==0].ID_IDADE, color='r', shade=True, Label= "CHURN")
plt.xlabel("IDADE")
plt.ylabel("Probability Density")
```

```python
sns.kdeplot(df.loc[df['Target']==1].VR_PLANO, color='b', shade=True, Label= "CHURN")
sns.kdeplot(df.loc[df['Target']==0].VR_PLANO, color='r', shade=True, Label= "CHURN")
plt.xlabel("Valor Plano")
plt.ylabel("Probability Density")
```

```python
sns.kdeplot(df.loc[df['Target']==1].TENURE, color='b', shade=True, Label= "CHURN")
sns.kdeplot(df.loc[df['Target']==0].TENURE, color='r', shade=True, Label= "CHURN")
plt.xlabel("Tempo de Casa")
plt.ylabel("Probability Density")
```

```python
sns.kdeplot(df.loc[df['Target']==1].SERASA, color='b', shade=True, Label= "CHURN")
sns.kdeplot(df.loc[df['Target']==0].SERASA, color='r', shade=True, Label= "CHURN")
plt.xlabel("SERASA")
plt.ylabel("Probability Density")
```

```python
colunas = df.columns
for x in colunas:
  print(x)
  if x=="Target":
    print("veio")
    next
  sns.kdeplot(df.loc[df['Target']==1][x], color='b', shade=True, Label= "CHURN")
  sns.kdeplot(df.loc[df['Target']==0][x], color='r', shade=True, Label= "CHURN")
  plt.xlabel(x)
  plt.ylabel("Probability Density")
```
